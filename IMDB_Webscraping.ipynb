```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import re
```

## Data scraping from the website

```python
movie_name = []
year = []
time = []
ratings_=[]
metascore = []
votes = []
gross = []
Release=[]
movie_type=[]
rate_=[]
for i in range(1,700,100):
    URL = 'https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start='+str(i)+'&ref_=adv_nxt'
    page = requests.get(URL)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    list_1 = soup.find_all('div', class_='lister-item-content')
    
    
    for store in list_1:
        year_of_release =store.find('h3', class_='lister-item-header')
        if year_of_release is None:
            year.append(np.NaN)
        else:
            year.append(year_of_release.text)
            
        ratings=store.find('span',class_='runtime')
        if ratings is None:
            ratings_.append(np.NaN)
        else:
            ratings_.append(ratings.text)
            
        meta=store.find('div',class_='inline-block ratings-metascore')
        if meta is None:
            metascore.append(np.NaN)
        else:
            metascore.append(meta.text)
        
        vote=store.find('p',class_='sort-num_votes-visible')
        if vote is None:
            votes.append(np.NaN)
        else:
            votes.append(vote.text)
            
        year_=store.find('span',class_='lister-item-year text-muted unbold')
        if year_ is None:
            Release.append(np.NaN)
        else:
            Release.append(year_.text)
            
        type_=store.find('span',class_='genre')
        if type_ is None:
            movie_type.append(np.NaN)
        else:
            movie_type.append(type_.text)
        
        rate=store.find('div',class_='inline-block ratings-imdb-rating')
        if rate is None:
            rate_.append(np.NaN)
        else:
            rate_.append(rate.text)
```

```python
len(year)
```

```python
movie= pd.DataFrame({'Movie_Name': year,'Time_mnts':ratings_,'Metascore':metascore,'Source':votes,'Release_Year':Release,'Movie_type':movie_type,'Ratings':rate_})
```

### Cleansing

```python
movie['Ratings']=movie['Ratings'].apply(lambda x:re.search('\d+\.?\d+',x).group()).astype(float)
```

```python
movie['Time_mnts']=movie['Time_mnts'].apply(lambda x:re.search('\d+',x).group()).astype(int)
```

```python
movie['Release_Year']=movie['Release_Year'].apply(lambda x:re.search('\d+',x).group()).astype(int)
```

```python
movie['Movie_type']=movie['Movie_type'].apply(lambda x:re.search('\s\D+\,?\D+\,?\D+',x).group().replace('\n',''))
```

```python
movie['Votes']=movie['Source'].apply(lambda x:re.search('\d+\,?\d+\,?\d+',x).group().replace(',','')).astype(int)
```

```python
movie
```

```python
movie['Trail_9']=movie['Source'].apply(lambda x:re.findall('\d+\.?\d+M',x))
```

```python
movie['Gross']=movie['Trail_9'].apply(lambda x: np.nan if len(x)==0 else x)
```

```python
movie.drop('Trail_9',axis=1,inplace=True)
```

```python
movie.drop('Source',axis=1,inplace=True)
```

```python
movie['Movie_Name']=movie['Movie_Name'].apply(lambda x:re.search('\s\D+',x).group().replace('\n','').replace('(',''))
```

```python
movie
```

```python
movie.dropna(inplace=True)
```

```python
movie['Metascore']=movie['Metascore'].apply(lambda x:re.search('\d+',x).group()).astype(int)
```

```python
movie
```

```python
movie=movie.reindex(columns=['Movie_Name','Release_Year','Time_mnts','Movie_type','Ratings','Metascore','Votes','Gross'])
```

```python
movie
```

```python
movie['Gross']=movie.apply(lambda x: pd.Series(x['Gross']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
movie['Gross_$']=movie.Gross.apply(lambda x: float(x.strip('M'))*1000000).astype(int)
```

```python
movie.drop('Gross',axis=1,inplace=True)
```
