```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import re
```

## Data scraping from the website

```python
baseurl="https://www.bikes4sale.in"
url_list=[]
for x in range(1,20):
    URL="https://www.bikes4sale.in/used/buy/?price=40000-60000:60000-80000:80000-100000:100000-200000:200000-9999999&ccrange=140-170:170-215:215-280:280-380:380-9999&page="+str(x)
    page=requests.get(URL)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    productlist=soup.find_all("div",class_="txtBox")
    for i in productlist:
        for link in i.find_all("a",href=True):
            url_list.append(baseurl+link['href'])
    #print("Original : ",i.find_all("a"))
    #a = re.findall(r'href=\"(.*)\"\starget',str(i))
    #final = "https://www.bikes4sale.in"+a[0]
    #print(final)
len(url_list)  
```

```python
data_list=[]
for link in url_list:
    page=requests.get(link)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
  
    Model=soup.find_all("table",class_="b4stable itemDtlsTbl")[0].find_all('td')[1].text
    Buy_Year=soup.find_all("table",class_="b4stable itemDtlsTbl")[0].find_all('td')[3].text
    Location=soup.find_all("table",class_="b4stable itemDtlsTbl")[0].find_all("td")[5].text
    Engine_CC=soup.find_all("table",class_="table idNewTbl")[0].find_all("td")[1].text
    No_of_Owners=soup.find_all("table",class_="table idNewTbl")[0].find_all("td")[3].text
    KMs_Run=soup.find_all("table",class_="table idNewTbl")[-2].find_all('tr')[-1].find_all('td')[1].text
    if len(soup.find_all("table",class_="table idNewTbl")[-1].find_all('tr'))>=2:
        Date_of_posting=soup.find_all("table",class_="table idNewTbl")[-1].find_all('tr')[-2].find_all('td')[1].text
    else:
        Date_of_posting = "Mosha"
    print(Date_of_posting)
    price=soup.find_all("div",class_="specLnk specLnkEmi serLnks")[0].find_all("b")[0].text
    data = {"Model":Model,"Buy_Year":Buy_Year,"Location":Location,"Engine_CC":Engine_CC,"No_of_Owners":No_of_Owners,"KMs_Run":KMs_Run,"price":price,"Date_of_posting":Date_of_posting}
    data_list.append(data)
  
```

```python
bike=pd.DataFrame(data_list)
```

```python
bike['price']=bike['price'].apply(lambda x:re.search('\s\d+\,?\d+\,?\d+',x).group().replace(',','')).astype(int)
```

```python
bike['Engine_CC']=bike['Engine_CC'].apply(lambda x:re.search('\d+',x).group()).astype(int)
```

```python
bike['Brand']=bike['Model'].apply(lambda x:re.search('\w+',x).group())
```

```python
for i in bike.index:
    if (bike.loc[i,'Brand']=='Royal'):
        bike.loc[i,'Brand']='Royal Enfield'
    elif (bike.loc[i,'Brand']=='Harley'):
        bike.loc[i,'Brand']='Harley Davidson'
    elif (bike.loc[i,'KMs_Run']=='Single Owner'):
        bike.loc[i,'KMs_Run']=np.NaN
```
